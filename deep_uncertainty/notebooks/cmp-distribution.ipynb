```python
import numpy as np
from scipy.special import factorial
import matplotlib.pyplot as plt
```

```python
def get_z(mu, phi, max_y=30):
    Z = 0
    for j in range(max_y):
        left = np.power((mu + (np.exp(phi) - 1)/(2*np.exp(phi))), j*np.exp(phi))
        right = 1/(np.power(factorial(j), np.exp(phi)))
        Z += left * right
    return Z
```

```python
def cmp_pmf(mu, phi, y, max_y=30):
    Z = get_z(mu, phi,max_y=max_y)
    exponent = y*np.exp(phi)
    left = np.power((mu + (np.exp(phi) - 1)/(2*np.exp(phi))), exponent)
    right = np.power((factorial(y)), -np.exp(phi))/Z
    
    return left * right
```

```python
def log_cmp_pmf(mu, phi, y):
    first = y*np.exp(phi)*np.log(mu + (np.exp(phi) - 1)/(2*np.exp(phi)))
    second = -1*np.exp(phi)*np.log(factorial(y))
    Z = get_z(mu, phi)
    third = np.log(Z)
    return first * second * third
```

## mu = 2

```python
y = np.arange(0, 25)
```

```python

phi_0 = -0.7
phi_1 = 0.0
phi_2 = 1.0
```

```python
probs_0 = cmp_pmf(mu=2, phi=phi_0, y=y)
probs_1 = cmp_pmf(mu=2, phi=phi_1, y=y)
probs_2 = cmp_pmf(mu=2, phi=phi_2, y=y)
```

```python
plt.plot(y, probs_0, linestyle='--', marker='o', label=f'phi={phi_0}')
plt.plot(y, probs_1, linestyle='--', marker='o', label=f'phi={phi_1}')
plt.plot(y, probs_2, linestyle='--', marker='o', label=f'phi={phi_2}')
plt.legend()
plt.show()
```

```python
probs_2.round(2)
```

## Normalizing constant

```python
mu = 15
phi = 0.7
max_y = 100
```

```python
Z = get_z(mu, phi, max_y)
```

```python
Z_arr=[]
Z = 0
for j in range(max_y):
    left = np.power((mu + (np.exp(phi) - 1)/2*np.exp(phi)), j*np.exp(phi))
    right = 1/(np.power(factorial(j), np.exp(phi)))
    Z += left * right
    Z_arr.append(Z)
```

```python
plt.title("Infinte Series in Z, normalizing constant")
plt.plot(np.arange(0, max_y), Z_arr)
plt.xlabel("y")
plt.ylabel("Z(mu, phi)")
plt.show()
```
