```python
from models.random_variables import GammaCount
import numpy as np
from matplotlib import pyplot as plt
from itertools import product
```

```python
logalpha_vals = np.linspace(-1, 1, num=5)
logbeta_vals = np.linspace(-0.5, 2, num=5)

fig, axs = plt.subplots(5, 5, figsize=(10, 10))    
for ax, (logalpha, logbeta) in zip(axs.ravel(), product(logalpha_vals, logbeta_vals)):
    rv = GammaCount(np.exp(logalpha), np.exp(logbeta))

    support = np.arange(0, 15)
    ax.set_title(f"$log(\\alpha) = {logalpha}, log(\\beta) = {logbeta}$", fontsize=6)
    ax.plot(support, rv.pmf(support))
    ax.set_xticks(support[::2])
    ax.set_ylim(0, 1)

plt.tight_layout()
plt.show()
```
